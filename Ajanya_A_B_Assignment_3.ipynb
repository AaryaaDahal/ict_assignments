1. Write a Python function to ﬁnd the Max of three numbers

```python
def max_of_three( x, y ,z):
    if (x > y) and (x > z):
        return  x
    elif (y > x) and (y > z):
        return  y
    else:
        return  z
```

```python
x=int(input("enter the first number: "))
y=int(input("enter the second number: "))
z=int(input("enter the third number: "))
print('max_of_three is',max_of_three(x,y,z))
```

2. Write a Python function to sum all the numbers in a list.  Sample List : [18, 12, 30,10, 7] Expected Output : 77

```python
def sum_list(items):
    s = 0
    for x in items:
        s+=x
    return s
```

```python
list=[18,12,30,10,7]
print('Sum is ',sum_list(list))
```
